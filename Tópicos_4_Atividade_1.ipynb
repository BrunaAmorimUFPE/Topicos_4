```python
 # # 1. Comentários, Indentação e Operadores

# Em Python, os comentários são criados com o caractere '#'. Tudo o que vem após o '#' é ignorado pelo interpretador.
# Comentários são úteis para explicar o código ou desabilitar partes temporariamente.

# Exemplo de comentário:
# Este é um comentário

# Indentação:
# Python usa indentação (espaços ou tabulações) para definir blocos de código, em vez de chaves como outras linguagens.
# A indentação é crucial para que o código funcione corretamente.

# Operadores:
# - Operadores Aritméticos: +, -, *, /, // (divisão inteira), % (módulo), ** (potência)
# - Operadores de Comparação: ==, !=, <, >, <=, >=
# - Operadores Lógicos: and, or, not

# Exemplo de código:

a = 10
b = 5

# Operações Aritméticas
print(a + b)  # Soma
print(a - b)  # Subtração
print(a * b)  # Multiplicação
print(a / b)  # Divisão
print(a // b) # Divisão inteira
print(a % b)  # Módulo
print(a ** b) # Potência

# Comparações
print(a == b) # Igualdade
print(a != b) # Desigualdade
print(a > b)  # Maior que
print(a < b)  # Menor que

# Lógicos
print(a > 0 and b < 10)  # and
print(a > 0 or b < 10)   # or
print(not(a > 0))        # not


# # 2. Tipos de Dados

# Python tem alguns tipos de dados principais:

# - Inteiros (int): Números inteiros, sem ponto decimal.
# - Flutuantes (float): Números com ponto decimal.
# - Complexos (complex): Números complexos, representados por "a + bj".
# - Strings (str): Sequências de caracteres, representadas por aspas simples ou duplas.
# - Booleanos (bool): Valores lógicos: True ou False.

# Exemplos:

inteiro = 10
flutuante = 3.14
complexo = 2 + 3j
texto = "Olá, Python!"
booleano = True

# Exibindo os tipos de dados
print(type(inteiro))    # <class 'int'>
print(type(flutuante))  # <class 'float'>
print(type(complexo))   # <class 'complex'>
print(type(texto))      # <class 'str'>
print(type(booleano))   # <class 'bool'>


# # 3. Estruturas de Dados

# As principais estruturas de dados em Python são:

# - Listas: São mutáveis (podem ser alteradas) e podem conter elementos de diferentes tipos.
# - Tuplas: São imutáveis (não podem ser alteradas) e podem conter diferentes tipos de elementos.
# - Conjuntos: São coleções não ordenadas e não permitem elementos duplicados.
# - Dicionários: São coleções de pares chave-valor, onde cada chave é única.

# Exemplos:

# Listas
lista = [1, 2, 3, "Python"]
lista.append(4)  # Adicionando um item
print(lista)

# Tuplas
tupla = (1, 2, 3, "Python")
print(tupla)

# Conjuntos
conjunto = {1, 2, 3, 4, 5}
conjunto.add(6)  # Adicionando um item
conjunto.remove(4)  # Removendo um item
print(conjunto)

# Dicionários
dicionario = {"nome": "João", "idade": 25}
print(dicionario)
dicionario["cidade"] = "São Paulo"  # Adicionando um novo par chave-valor
print(dicionario)


# # 4. Condicionais e Loops

# Condicionais (if, elif, else):
# Permitem executar blocos de código diferentes com base em uma condição.

x = 10
if x > 5:
    print("x é maior que 5")
elif x == 5:
    print("x é igual a 5")
else:
    print("x é menor que 5")

# Loops:
# - for: Usado para iterar sobre uma sequência (como uma lista, tupla ou intervalo).
# - while: Executa um bloco de código enquanto uma condição for verdadeira.

# Exemplo de loop for:
for i in range(5):  # Vai de 0 até 4
    print(i)

# Exemplo de loop while:
i = 0
while i < 5:
    print(i)
    i += 1  # Incrementa 1 a cada iteração


# # 5. Funções

# Funções são blocos de código que podem ser reutilizados em diferentes partes do programa.
# São definidas usando a palavra-chave 'def'.

def saudacao(nome):
    return f"Olá, {nome}!"

# Chamando a função
print(saudacao("João"))

# Funções Lambda:
# São funções anônimas, geralmente usadas para tarefas simples, definidas com a palavra-chave 'lambda'.

soma = lambda x, y: x + y
print(soma(3, 4))


# # 6. Trabalhando com Arquivos

# Python possui a função 'open()' para abrir arquivos. É possível abrir arquivos para leitura, escrita ou anexação.
# O comando 'with' é utilizado para garantir que o arquivo será fechado após o uso.

# Abrindo e lendo um arquivo:
with open("exemplo.txt", "r") as file:
    conteudo = file.read()
    print(conteudo)

# Abrindo e escrevendo em um arquivo:
with open("exemplo.txt", "w") as file:
    file.write("Olá, este é um exemplo de escrita no arquivo.")

# O arquivo 'exemplo.txt' será criado ou sobrescrito com o conteúdo acima.


# # 7. Importando Módulos

# Python vem com muitos módulos embutidos, como o 'math' (para funções matemáticas) e 'random' (para números aleatórios).
# Você pode importar módulos com a palavra-chave 'import'.

import math
import random

# Usando o módulo math:
print(math.sqrt(16))  # Raiz quadrada de 16

# Usando o módulo random:
print(random.randint(1, 10))  # Gera um número aleatório entre 1 e 10


# # 8. Tratamento de Erros

# O bloco 'try...except' é usado para capturar e tratar exceções (erros) durante a execução do programa.
# Isso evita que o programa pare de funcionar devido a um erro inesperado.

try:
    resultado = 10 / 0  # Tentando dividir por zero (erro)
except ZeroDivisionError:
    print("Erro: Não é possível dividir por zero.")

# Também é possível usar 'else' e 'finally' para capturar erros e executar ações adicionais.

try:
    resultado = 10 / 2
except ZeroDivisionError:
    print("Erro: Não é possível dividir por zero.")
else:
    print(f"Resultado: {resultado}")  # Será executado se não houver erro
finally:
    print("Bloco try-except concluído.")  # Sempre será executado


```
